```python
import os
import getpass
import pandas as pd
from sqlalchemy import create_engine, text
```

```python
os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter OpenAI API Key: ")
```

```python
if os.path.exists("analytics.db"):
    os.remove("analytics.db")

engine = create_engine("sqlite:///analytics.db")

print("SQLite engine created")
```

```python
users = pd.DataFrame({
    "user_id": [1,2,3,4,5],
    "customer_name": ["Amit","Priya","Rahul","Sneha","Arjun"],
    "region": ["East","West","North","South","East"]
})

sales = pd.DataFrame({
    "sale_id": [101,102,103,104,105,106,107],
    "user_id": [1,2,3,4,5,1,3],
    "amount": [500,700,300,900,650,400,800]
})

print("Dummy data ready")
```

```python
with engine.begin() as conn:
    conn.execute(text("DROP TABLE IF EXISTS users"))
    conn.execute(text("DROP TABLE IF EXISTS sales"))

    conn.execute(text("""
        CREATE TABLE users (
            user_id INTEGER,
            customer_name TEXT,
            region TEXT
        )
    """))

    conn.execute(text("""
        CREATE TABLE sales (
            sale_id INTEGER,
            user_id INTEGER,
            amount INTEGER
        )
    """))

    for _, row in users.iterrows():
        conn.execute(
            text("INSERT INTO users VALUES (:user_id, :customer_name, :region)"),
            {
                "user_id": int(row["user_id"]),
                "customer_name": row["customer_name"],
                "region": row["region"]
            }
        )

    for _, row in sales.iterrows():
        conn.execute(
            text("INSERT INTO sales VALUES (:sale_id, :user_id, :amount)"),
            {
                "sale_id": int(row["sale_id"]),
                "user_id": int(row["user_id"]),
                "amount": int(row["amount"])
            }
        )

print("Tables created successfully")
```

```python
from langchain_community.utilities import SQLDatabase

db = SQLDatabase(engine)

print(db.get_usable_table_names())
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)

print("OpenAI LLM ready")
```

```python
from langchain_community.agent_toolkits import SQLDatabaseToolkit

toolkit = SQLDatabaseToolkit(db=db, llm=llm)
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name)
```

```python
from langchain_community.agent_toolkits import create_sql_agent

agent_executor = create_sql_agent(
    llm=llm,
    db=db,
    agent_type="tool-calling",
    verbose=True
)

print("SQL Agent recreated")
```

```python
response = agent_executor.invoke({
    "input": "Top 5 customers by revenue"
})

print(response["output"])
```

```python
response = agent_executor.invoke({
    "input": "Show total revenue by region"
})

print(response["output"])
```
